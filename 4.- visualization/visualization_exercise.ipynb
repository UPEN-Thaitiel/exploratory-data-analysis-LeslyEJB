### Visualization

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

#### Create Line Plot

```python
years = [2010, 2011, 2012, 1013]
sales = [100, 120, 140, 160]
```

```python
plt.plot(years, sales, label="Sales Trend", color='blue', marker='o')
plt.title("Sales over the years")
plt.xlabel(years)
plt.ylabel(sales)
plt.legend()
plt.show()
```

#### Bar Chart

```python
categories = ["Electronics", "Clothing", "Furniture"]
revenue = [250, 400, 150]
```

```python
plt.bar(categories, revenue, color="green")
plt.show()
```

#### Scatter Plot

```python
hours_studied = [1, 2, 3, 4, 5]
exam_score = [50, 55, 67, 80, 90]
```

```python
plt.scatter(hours_studied, exam_score, color="black")
plt.title("Study hours vs examn score")
plt.xlabel("hours")
plt.ylabel("score")
plt.show()
#histogram
```
